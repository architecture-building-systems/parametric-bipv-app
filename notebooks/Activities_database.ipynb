```python
def calculate_params(material, location, unit_per_m2):
    materials = db.search(material)
    x = None
    for m in materials:
        if m.get("location") == location:
            x = m
            break
    if x is None:
        return {}
    act = x.key
    method_key = []
    for m in bw.methods:
    #    if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m):
    #        if 'climate change' in str(m):
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'climate change' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'terrestrial acidification' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'freshwater eutrophication' in str(m) and not 'no LT' in str(m):
            method_key.append(m) 
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'marine eutrophication' in str(m) and not 'no LT' in str(m):
            method_key.append(m)   
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'metal depletion' in str(m) and not 'no LT' in str(m):
            method_key.append(m)   
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'particulate matter formation' in str(m) and not 'no LT' in str(m):
            method_key.append(m) 
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'ozone depletion' in str(m) and not 'no LT' in str(m):
            method_key.append(m)  
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'photochemical oxidant formation' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'agricultural land occupation' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'natural land transformation' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'urban land occupation' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'freshwater ecotoxicity' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'human toxicity' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'ionising radiation' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'ReCiPe Midpoint (H) V1.13' in str(m) and  'Midpoint (H)' in str(m) and 'water depletion' in str(m) and not 'no LT' in str(m):
            method_key.append(m)
        if 'cumulative energy demand' in str(m) and  'fossil' in str(m) and 'non-renewable energy resources, fossil' in str(m):
            method_key.append(m) 
    functional_unit = {act:1}
    params = {}    
    for m in method_key:    
        lca = bw.LCA(functional_unit,m)
        lca.lci()
        lca.lcia()
        params[m[1]] = lca.score*unit_per_m2
    return params

        
```
